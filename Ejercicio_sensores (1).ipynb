```
lineas_medidas = sc.textFile(File)

lineas_medidas.take(4)
```

```
medidas_cadena = lineas_medidas.flatMap(lambda elemento: elemento.split(','))

medidas_cadena.take(10)
```

```
medidas_numero = medidas_cadena.map(lambda elemento: float(elemento))

medidas_numero.take(10)
```

```
medidas_numero.filter(lambda elemento: elemento<0).count()
```

```
medidas_numero.filter(lambda elemento: elemento<0).collect()
```

```
medidas_correctas = medidas_numero.filter(lambda elemento: elemento>=0)

medidas_correctas.filter(lambda elemento: elemento<0).count()
```

```
suma = medidas_correctas.reduce(lambda e1, e2: e1 + e2)

cantidad_medidas = medidas_correctas.count()

print('La media es: ', suma/cantidad_medidas)
```
